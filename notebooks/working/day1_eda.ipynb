```python
# Check missing values
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Hidden missing values check for TotalCharges
print("Blank TotalCharges entries:", (df['TotalCharges'] == ' ').sum())
```

### Observation
- `TotalCharges` contains blank string values even though it should represent a numeric billing field.
- This is a hidden missing-value issue, so the column will need cleaning before preprocessing and modeling.
- Catching this early improves data quality and shows attention to detail in the EDA phase.

```python
# Check duplicate rows
print('Duplicate Rows:', df.duplicated().sum())
```

```python
# Target distribution: Churn
plt.figure(figsize=(6,4))
sns.countplot(x='Churn', data=df)
plt.title('Customer Churn Distribution')
plt.xlabel('Churn')
plt.ylabel('Count')
plt.show()
```

### Observation:
- The dataset appears slightly imbalanced: more `No` churn than `Yes`.
- This means we should consider appropriate evaluation metrics and possibly resampling or class-weighted models later.
- Business note: even a small churn percentage can correspond to large revenue loss, so recall on the churn class will be important.

```python
# Churn percentage
churn_percentage = df['Churn'].value_counts(normalize=True) * 100
print(churn_percentage)
```

```python
# Numerical vs Categorical features
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = df.select_dtypes(include=['object']).columns.tolist()
print('Numerical Features:', numerical_features)
print('Categorical Features:', categorical_features)
```

## Customer Demographics: Gender vs Churn

```python
plt.figure(figsize=(6,4))
sns.countplot(x='gender', hue='Churn', data=df)
plt.title('Gender vs Churn')
plt.show()
```

### Observation:
- Churn distribution appears relatively similar across genders.
- Gender alone may not be a strong predictor of churn, but it could interact with other features.
- Business interpretation: retention efforts likely shouldn't be gender-targeted without further segmentation.

## Senior Citizen vs Churn

```python
plt.figure(figsize=(6,4))
sns.countplot(x='SeniorCitizen', hue='Churn', data=df)
plt.title('Senior Citizen vs Churn')
plt.show()
```

### Observation:
- Senior citizens appear to have a higher proportion of churn compared to non-senior customers.
- Possible reasons: product fit, pricing sensitivity, or service usability.
- Business recommendation: investigate tailored support or simpler plans for senior customers.

## Contract Type vs Churn

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title('Contract Type vs Churn')
plt.xticks(rotation=10)
plt.show()
```

### Observation:
- Customers with Month-to-month contracts have significantly higher churn rates.
- Longer contracts (One year, Two year) show lower churn — suggesting contract length increases retention.
        - Business implication: incentivize longer contracts with discounts or added benefits.

```python
# Churn percentage by contract type
contract_churn_pct = pd.crosstab(df['Contract'], df['Churn'], normalize='index') * 100
contract_churn_pct.round(2)
```

## Payment Method vs Churn

```python
plt.figure(figsize=(10,5))
sns.countplot(x='PaymentMethod', hue='Churn', data=df)
plt.title('Payment Method vs Churn')
plt.xticks(rotation=20)
plt.show()
```

### Observation:
- Payment methods using electronic checks often show higher churn proportions in this dataset.
- This could indicate failed payments, billing friction, or differences in customer segments.
- Business action: review e-check payment flows and consider targeted retention for customers on e-checks.

## Tenure Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['tenure'], bins=30, kde=True)
plt.title('Customer Tenure Distribution')
plt.xlabel('Tenure (Months)')
plt.show()
```

### Observation:
- Tenure distribution is skewed toward newer customers (shorter tenure).
- Short-tenure customers are more likely to churn (confirm with aggregation later).
- Business takeaway: onboarding and early engagement are critical for retention.

## Monthly Charges vs Churn

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title('Monthly Charges vs Churn')
plt.show()
```

### Observation
- Customers with higher monthly charges tend to churn more often.
- This may indicate price sensitivity or dissatisfaction among higher-value customers.
- MonthlyCharges should be considered an important predictive feature and a useful retention segmentation variable.
